Estudio de factores determinantes en casos de depresión

El dataset selecionado muestra información relacionada con el diagnóstico de  depresión en un grupo de estudiantes.  La data contiene variables como la edad, género y  ciudad que permiten identificar a los individuos, además de otras  que muestran factores externos posiblemente influyentes como el tiempo de sueño, satisfacción de estudio, presión laboral,habitos de la dieta satisfacción laboral, estress financiero y realción trabajo estudio.  Adicionalmente  se tiene datos directamente asociados a la enfermedad como antecedentes familiares y pensaminetos suicidas.
Lo que se quiere es buscar posibles relaciones directas en entre las vaeiables a través de correlaciónes a través de gráfico.
En el estudio se observarán las cantidades de estudiantes deprimidos junto  variables identificadoras como edad y género. Igualmente se detallan los indicadores de pensaminetos suicidas, historia familiar de problemas mentales y duración del sueño. Todo ello con la finalidad de responder a   
Las interrogantes plateandas son las siguientes:
¿Cuál es la relación edad-depresión?
¿Cuando se presentan mas pensamientos suicidas con respecto a la edad?
¿Como los pensamientos suicidas estan directamente relacionados con la depresión?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url= 'https://raw.githubusercontent.com/LeoV79080679/DSCI/refs/heads/main/student_depression_dataset.csv'

df=pd.read_csv(url)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
total_hombres = df['Gender'].value_counts()['Male']
total_mujeres = df['Gender'].value_counts()['Female']
print(f"Total hombres: {total_hombres}")
print(f"Total mujeres: {total_mujeres}")

Generos = ['Male', 'Female']

plt.figure(figsize=(8,6))
plt.bar(Generos, [total_hombres, total_mujeres], color=['blue', 'red'])

plt.xlabel('Género')
plt.ylabel('Número de Personas')
plt.title('Comparación entre Total de Hombres y Mujeres')

plt.show()
```

```python
depresivos = df[df['Depression'] == 1]
no_depresivos = df[df['Depression'] == 0]
len(depresivos)
len(no_depresivos)

print(f"Total depresivos: {len(depresivos)}")
print(f"Total no depresivos: {len(no_depresivos)}")
```

```python
Estados_depresion= ['depresivos', 'no_depresivos']
totales = [16336, 11565]

plt.figure(figsize=(8, 6))
plt.bar(Estados_depresion, totales, color=['red', 'blue'])

plt.xlabel('Estados de depresión')
plt.ylabel('Número de Personas')
plt.title('Comparación entre Total de Personas Depresivas y No Depresivas')

plt.show()
```

```python
df['Sleep Duration'].str.strip()

df['Sleep Duration'].unique()

sleep_duration= df['Sleep Duration'].value_counts()

print(sleep_duration)

plt.figure(figsize=(8, 6))
plt.bar(sleep_duration.index, sleep_duration.values, color='blue')

plt.xlabel('Horas de sueño')
plt.ylabel('Número de Personas')
plt.title('Distribución de Horas de Sueño')

plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.boxplot(df['Age'])
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Años')
plt.show()
```

```python
df['Have you ever had suicidal thoughts ?'].unique()
ideas_suicidas = df['Have you ever had suicidal thoughts ?'].value_counts()
print(ideas_suicidas)

plt.figure(figsize=(8,6))
sns.countplot(df['Have you ever had suicidal thoughts ?'])
plt.title('Relacion de Ideas Suicidas')
plt.xlabel('cantidad')
plt.ylabel('opciones')
plt.show()
```

```python
df['Family History of Mental Illness'].unique()
historial_mental = df['Family History of Mental Illness'].value_counts()
print(historial_mental)

sns.countplot(df['Family History of Mental Illness'])
plt.title('historia familiar de problemas mentales')
plt.xlabel('cantidad')
plt.ylabel('opciones')
plt.show()
```

```python
plt.figure(figsize=(15,10))

sns.countplot(x=df['Age'], hue=df['Depression'])
plt.title('Relacion entre Edad y Depresion')
plt.xlabel('Edad')
plt.ylabel('Depresion')

```

En la relación Edad-Depresión (Depresivo=1, no depresivo=0) se puede observar que la la edad juega un factor determinante. A menor edad ha una marcada presencia de estados de depresión, hecho que contrasta con edades mayores a 30 años dentro de la población estudiantil.  

```python
plt.figure(figsize=(15,10))

sns.countplot(x=df['Age'], hue=df['Have you ever had suicidal thoughts ?'])
plt.title('Relacion entre Edad y pensamientos sucidas')
plt.xlabel('Edad')
plt.ylabel('Pensaminetos suicidas')
plt.legend(title='Pensamientos suicidas', loc='upper right')
plt.show()
```

Si bien los pensamientos suicidas estan presentes en todas las edades, en estudiantes menores  a 30 años la cantidad es de casi el doble de aquellos sin pensamientos suicidas.  

```python
print(f"Las opciones de historia de enfermedad mental son: {df['Family History of Mental Illness'].unique()}")


print(f"Las opciones de historia de enfermedad mental son: {df['Family History of Mental Illness'].unique()}")


print(f" Las cantidad de cada opción son: {df['Family History of Mental Illness'].value_counts()}")

plt.figure(figsize=(15,10))
sns.countplot(x=df['Depression'], hue=df['Family History of Mental Illness'])
plt.title('Relacion entre la depresión y la historia de enfermedad mental')
plt.xlabel('depresión')
plt.ylabel('Historia de enfermedad mental')
plt.legend(title='Historia de enfermedad mental', loc='upper left')
```

En este gráfico se puede observar como los antecedentes familiares puedieran ser influyentes en los casos de depresión, ello esta  muy marcado en la información contrastada en el número de estudiantes mostrada a través de la comparación de los gráficos de barra.
