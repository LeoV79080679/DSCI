Estudio de factores determinantes en casos de depresión

El dataset selecionado muestra  estudiantes con depresión o no,  observandose variables que identifican a cada sujeto junto a factores externos asociados al estado depresivo. Lo que se busca en este estudio es establecer la relacion de estos factores con la depresión y la posible influencia en el diagnóstico.

Las interrogantes plateandas son las siguientes:
¿Puede es el género estas asociado a la depresión?
¿Cual es la influencia de la presión académica sobre la depresión?
¿Los pensamientos suicidas estan directamente relacionados con al inflación?



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url= 'https://raw.githubusercontent.com/LeoV79080679/DSCI/refs/heads/main/student_depression_dataset.csv'

df=pd.read_csv(url)
```

```python
total_hombres = df['Gender'].value_counts()['Male']
total_mujeres = df['Gender'].value_counts()['Female']
print(f"Total hombres: {total_hombres}")
print(f"Total mujeres: {total_mujeres}")

Generos = ['Male', 'Female']

plt.figure(figsize=(8,6))
plt.bar(Generos, [total_hombres, total_mujeres], color=['blue', 'red'])

plt.xlabel('Género')
plt.ylabel('Número de Personas')
plt.title('Comparación entre Total de Hombres y Mujeres')

plt.show()
```

```python
depresivos = df[df['Depression'] == 1]
no_depresivos = df[df['Depression'] == 0]
len(depresivos)
len(no_depresivios)

print(f"Total depresivos: {len(depresivos)}")
print(f"Total no depresivos: {len(no_depresivos)}")
```
